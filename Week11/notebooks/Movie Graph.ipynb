```python
def get_content(file='./data/The.Matrix.1999.720p.BluRay.264.YIFY.txt'):
    names = []
    last_name = ''
    
    with open(file) as f:
        for line in f.readlines():
            if '-->' in line:
                continue
            
            if ':' in line:
                line = line.replace('<i>', '')
                part = line.split(':')[0]
                
                if not any([ x.islower() for x in part ]):
                    if last_name != part:
                        names.append( part )
                        last_name = part
                    
                
        return names
```

```python
names = get_content()
```
