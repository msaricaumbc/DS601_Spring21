```python
def get_all_files(location='./text_files/', file_types=['.txt', '.docx']):
    paths = []
    for root, dirs, files in os.walk(location):
        for file in files:
            if any([file.endswith(ft) for ft in file_types]):
                file_path =  os.path.join(root, file)
                # print(file_path)
                # print(file)
                paths.append( (file_path, file ) )  # I'm adding as a tupple (full path, file name) so that I don't need to parse file name separately
    # print(paths)
    return paths
```

```python
def read_doc(file):
    """
    reads doc/docx files

    return file content as a string
    """
    # https://www.pythonprogramming.in/how-to-read-data-from-docx-file-in-python.html
    try:
        doc = docx.Document(file)  # Creating word reader object.
        data = ""
        fullText = []
        for para in doc.paragraphs:
            fullText.append(para.text)
            data = ' '.join(fullText)

        return data

    except IOError:
        print('There was an error opening the file!')
        return ""
    
# read_doc('./text_files/week_10_document2.docx')
```

```python
def read_txt(file):
    """
    reads txt files 

    returns file content as a string
    """

    with open(file) as f:
        lines = f.readlines()
        return '\n'.join(lines)
    
# read_txt('./text_files/dr_yawn.txt')
```

```python
def save_obj(obj, file_name):
    with open(file_name, "w") as f:
        for key in obj.keys():
            line = key + ':\t' + ','.join(obj[key]) + '\n'
            f.write(line)
```

```python
def get_file_content(file):
    if '.doc' in file: 
        return read_doc(file)

    elif '.txt' in file: 
        return read_txt(file)

# get_file_content('./text_files/week_10_document2.docx') # sanity check
# get_file_content('./text_files/dr_yawn.txt') # sanity check
```

```python
def get_cleaned_words(content):
    # get stop words as a set
    stop_words_set = set(stopwords.words('english'))
    
    content = content.lower()
    
    # remove punctuation
    content = content.translate(str.maketrans('', '', string.punctuation))
    
    #split text into list of words
    words = word_tokenize(content)
  
    # remove duplicates and use sets so we can remove stop words set
    word_set = set(words)
     
    #remove stopwords
    word_set =  word_set - stop_words_set
    
    # convert back to list because we need to keep it as a list
    word_list = list(word_set)
    
    return  word_list

# get_cleaned_words('data science, Master Program is the best!!') # sanity check
```

```python
def extract_words_from_file(file):
    content = get_file_content(file)
    words = get_cleaned_words(content)
    return words

# extract_words_from_file('./text_files/week_10_document2.docx')
```

```python
obj = {}

for file_path, file in files:
    # print(file_path, file)
    words = extract_words_from_file(file_path)
    obj[file] = words
    
save_obj(obj,'output.dat')
```
